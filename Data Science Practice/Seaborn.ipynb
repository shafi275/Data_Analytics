```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your Titanic dataset
titanic_data = pd.read_csv("C:/Users/makmo/Downloads/titanic.csv")
```

```python
# Scatter Plot: Age vs. Fare
plt.figure(figsize=(8, 6))
sns.scatterplot(data=titanic_data, x='Age', y='Fare', hue='Survived', palette='viridis', alpha=0.7)
plt.title('Scatter Plot: Age vs. Fare by Survival')
plt.show()
```

```python
# Count Plot: Survival by Gender
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_data, x='Sex', hue='Survived', palette='pastel')
plt.title('Count Plot: Survival by Gender')
plt.show()
```

```python

# Bar Plot: Survival Rate by Class
plt.figure(figsize=(8, 6))
sns.barplot(data=titanic_data, x='Pclass', y='Survived', ci=None, palette='muted')
plt.title('Bar Plot: Survival Rate by Class')
plt.show()
```

```python
# Boxplot: Fare Distribution by Class
plt.figure(figsize=(8, 6))
sns.boxplot(data=titanic_data, x='Pclass', y='Fare', palette='coolwarm')
plt.title('Boxplot: Fare Distribution by Class')
plt.show()
```

```python
# Violin Plot: Age Distribution by Survival
plt.figure(figsize=(8, 6))
sns.violinplot(data=titanic_data, x='Survived', y='Age', palette='Set2', split=True)
plt.title('Violin Plot: Age Distribution by Survival')
plt.show()
```

```python
# Select only numeric columns for the correlation matrix
numeric_data = titanic_data.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap: Correlation Matrix')
plt.show()
```

```python
# Pairplot: Relationships Between Variables
sns.pairplot(titanic_data, vars=['Age', 'Fare', 'Survived'], hue='Survived', palette='husl')
plt.suptitle('Pairplot: Relationships Between Age, Fare, and Survival', y=1.02)
plt.show()
```

```python
# KDE Plot: Fare Distribution by Survival
plt.figure(figsize=(8, 6))
sns.kdeplot(data=titanic_data[titanic_data['Survived'] == 1], x='Fare', shade=True, label='Survived', color='green')
sns.kdeplot(data=titanic_data[titanic_data['Survived'] == 0], x='Fare', shade=True, label='Did Not Survive', color='red')
plt.title('KDE Plot: Fare Distribution by Survival')
plt.legend()
plt.show()
```
